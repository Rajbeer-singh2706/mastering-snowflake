```python
## 11. Setting up warehouse using SQL
'''
    create or replace warehouse SECOND_WH
    WITH 
    WAREHOUSE_SIZE = XS OR XSMALL
    AUTO_SUSPEND = 300 
    SCALING_POLICY = 'ECONOMY'
    MIN_CLUSTER_COUNT = 1
    MAX_CLUSTER_COUNT = 3
    INITIALLY_SUSPENDED = TRUE 
    COMMENT = 'This is our Second Warehouse'
'''
# CTRL+ENTER 
```
